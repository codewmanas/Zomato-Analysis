# Q.1] What type of restaurant do the majority of customers order from?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataframe = pd.read_csv("Zomato data .csv")

sns.countplot(x=dataframe['listed_in(type)']) # --> Counts the occurrences and plots accordingly
plt.xlabel("Type of Restaurant") # --> What to name the X-axis. The Y-axis has the default label 'count'
```

#### Conclusion - Majority of the restaurant falls in Dinning Category
